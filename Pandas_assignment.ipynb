```python
columns_to_cap = df['Sales']  

for Sales in columns_to_cap:
    Q1 = df['Sales'].quantile(0.25)
    Q3 = df['Sales'].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df['Sales'] = df['Sales'].clip(lower=lower_bound, upper=upper_bound)
```

# Average sales by state

```python
Avg_Sales_state = df.groupby("State")["Sales"].mean().to_frame().reset_index()
```

```python
sns.barplot(x='State', y = 'Sales', data  = Avg_Sales_state)
plt.show()
```

# Total quantity sold per category

```python
TotalQuant = df.groupby("Category")["Quantity"].sum().to_frame().reset_index()
```

```python
sns.barplot(x='Category', y= 'Quantity', data = TotalQuant)
plt.show()
```

# Subcategory with the highest average profit

```python
avg_sub_profit = df.groupby("SubCategory")["Profit"].mean().to_frame().reset_index()
sns.barplot(x= "SubCategory", y = "Profit", data = avg_sub_profit)
plt.show()
```
